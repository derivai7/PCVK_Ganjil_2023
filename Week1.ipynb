**Langkah 1**

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow #forImageDisplay
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

**Langkah 2**

```python
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://placekitten.com/800/571"]


for url in urls:
          image  = io.imread(url)
          image  = cv.resize(image, (0,0), fx=0.5, fy=0.5)
          image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
          final_frame = cv.hconcat ((image, image_2))
          cv2_imshow(final_frame)
          print('\n')
```

**Langkah 3**

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ", tinggi, " x ", lebar)
cv2_imshow(image_2)
```

**Langkah 4**

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (lebar):
  image_3[int((lebar)/3),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

**Tugas**

Buat garis vertikal dan garis menyilang diagonal pada image keluaran

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (tinggi):
  tengah = int(lebar/2)
  image_3[int(y),tengah] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis diagonal menyilang pada image
for i in range(lebar):
    y = int(i * (tinggi / lebar))
    image_3[y, i] = [255, 255, 255]
    image_3[y, lebar - i - 1] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

Buat garis horisontal berwarna putih dibagian tengah gambar dengan panjang tertentu

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (int(lebar/2)):
  image_3[int((lebar)/3),y] = [255,255,255]
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

Buat kotak menggunakan kumpulan pixel warna putih di sembarang tempat dalam gambar

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

kotak_x = 90
kotak_y = 90
lebar_kotak = 90
tinggi_kotak = 90

image_3[kotak_y:kotak_y + tinggi_kotak, kotak_x:kotak_x + lebar_kotak] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
